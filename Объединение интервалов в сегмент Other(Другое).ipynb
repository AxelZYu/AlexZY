```python
''' '''
import matplotlib.pyplot as plt
from matplotlib import ticker as ticker
import numpy as np

'''  '''
salares=[1215, 1221, 1263, 1267, 1271, 1274, 1318, 1320, 1324, 1324, 1326, 1337, 1346, 1354, 1367, 1372, 1375, 1376,
         1378, 1378, 1410, 1415, 1418, 1420, 1422, 1426, 1430, 1434, 1437, 1451, 1454, 1467, 1470, 1473, 1477, 1479,
         1480, 1514, 1516, 1522, 1529, 1544, 1547, 1554, 1562, 1584, 1595, 1616, 1626, 1717]
count, labels = np.histogram(salares, bins=np.arange(1100, 1900, 50))#     
'''     count  labels, count        
   labels      50   1100  1900'''

labels=['$'+str(labels[i])+'-'+str(labels[i+1]) for i , _ in enumerate(labels[1:])]#   labels       
'''       ,    2  .     '''
lst=list()
for i in count:
    if i <=2:
        lst.append(i)
'''      '''
non_min=[i for i, x in enumerate(count) if x>2]
count=[a for i, a in enumerate(count) if i in non_min]#    count
labels=[b for i, b in enumerate(labels) if i in non_min]#    labels
other=list()#  
other=sum(lst)#         count
count.append(other)#      count
'''  ,   count     labels,       labels'''
labels.append('Other')
'''      '''
plt.pie(count, labels=labels, autopct='%1.1f%%')
plt.title('    ')
plt.show()
```
