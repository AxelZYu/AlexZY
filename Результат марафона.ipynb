```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
import scipy.stats as stats
marafon=pd.read_csv("https://video.ittensive.com/python-advanced/marathon-data.csv", delimiter=",")

a=marafon['split'].apply(lambda x: x.split(':')[0])
a=a.astype(int)
b=marafon['split'].apply(lambda x: x.split(':')[1])
b=b.astype(int)
c=marafon['split'].apply(lambda x: x.split(':')[2])
c=c.astype(int)
marafon['split_sek']=a*3600+b*60+c
a=marafon['final'].apply(lambda x: x.split(':')[0])
a=a.astype(int)
b=marafon['final'].apply(lambda x: x.split(':')[1])
b=b.astype(int)
c=marafon['final'].apply(lambda x: x.split(':')[2])
c=c.astype(int)
marafon['final_sek']=a*3600+b*60+c

sns.set_context("paper", font_scale=2)
marafon1=pd.DataFrame(marafon, columns=["split_sek", "final_sek", "gender"])
sns.pairplot (marafon1, hue='gender', height=6)
plt.show()
print (round(stats.pearsonr(marafon1["split_sek"], marafon1['final_sek'])[0],2))
```
