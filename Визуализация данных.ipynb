```python
%matplotlib inline
from matplotlib import pyplot as plt
import pandas as pd
use=pd.read_csv("https://video.ittensive.com/python-advanced/data-9722-2019-10-14.utf.csv", delimiter=';')
use['AdmArea']=use['AdmArea'].apply(lambda x: x.split(' ')[0]).astype('category')
use=use.dropna(axis=1)
use=use.set_index('YEAR').loc['2018-2019']
use=use.reset_index()
use['District']=use['District'].apply(lambda x: x.split(' ')[1]).astype('category')
strog=use.set_index('District').loc['Строгино'].reset_index()
print ('Количество учеников написавших ЕГЭ более 220 балов:', end=' ')
print(strog['PASSES_OVER_220'].sum())

fig=plt.figure(figsize=(12, 12))
area=fig.add_subplot(2, 1, 1)
okrug=use.set_index('AdmArea')
over_220=okrug['PASSES_OVER_220'].groupby ('AdmArea').sum()

area.set_title("Отличники ЕГЭ")
over_220.plot.pie(ax=area, label="")
#Северо-запад Москва
czm=use.set_index('AdmArea').loc['Северо-Западный']
czm=czm.reset_index()
czm=czm.set_index('District')
czao=czm['PASSES_OVER_220'].groupby('District').sum()

czao=czao.reset_index()
non_zero=[i for i, x in enumerate(czao['PASSES_OVER_220']) if x!=0]
otlich=[a for i, a in enumerate(czao['PASSES_OVER_220']) if i in non_zero]

area=fig.add_subplot(2, 1, 2)
plt.title("СЗАО", fontsize=20)
plt.pie(otlich, labels=otlich, autopct='%1.1f%%')
area.legend(bbox_to_anchor=(1.6, 0.6))
plt.show()
```
